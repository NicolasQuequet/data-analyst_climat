```python
#On essaye de comprendre le lien entre co2_including_luc, primary_energy_consumption et co2_including_luc_per_unit_energy en faisant nous-même le calcul
tmp = df_france_20.loc[:,['country','year', 'gdp', 'co2_including_luc', 'co2_including_luc_per_unit_energy', 'primary_energy_consumption']]
tmp['co2_including_luc_per_unit_energy_bis'] = round(tmp['co2_including_luc']/tmp['primary_energy_consumption'],3)
tmp.head(20)
```

La cellule ci-dessus nous a permis de trouver la formule exacte de calcul liant co2_including_luc, co2_including_luc_per_unit_energy et	primary_energy_consumption.
Cette logique a été appliquée à d'autres formules pour comprendre d'autres relations (per_capita, per_gdp...) mais j'ai parfois exécuté le code dans la même cellule en le remplaçant. Cela m'a aidé à comprendre les relations entre les colonnes, indiquées dans le rapport, mais le code source n'a pas été conservé.

```python
#Ici j'essaie de comprendre comment est calculé co2 et co2_including_luc à partir des autres colonnes d'émission de co2 détaillées par sources.
tmp = df_france_20.loc[:,['country','year', 'co2_including_luc', 'co2', 'land_use_change_co2', 'cement_co2', 'coal_co2','consumption_co2', 'flaring_co2', 'gas_co2', 'oil_co2', 'other_industry_co2', 'trade_co2', 'total_ghg', 'total_ghg_excluding_lucf']]
tmp['co2 + luc_co2'] = tmp['co2'] + tmp['land_use_change_co2']
tmp['total_co2'] = tmp['cement_co2'] + tmp['coal_co2'] + tmp['flaring_co2'] + tmp['gas_co2'] + tmp['oil_co2'] + tmp['other_industry_co2'] + tmp['trade_co2']
tmp['total_co2_no_trade'] = tmp['cement_co2'] + tmp['coal_co2'] + tmp['flaring_co2'] + tmp['gas_co2'] + tmp['oil_co2'] + tmp['other_industry_co2']

display(tmp)
```

Grâce au code ci-dessus, j'en déduis que co2 est la somme des colonnes détaillées de co2 (ciment_co2, coal_co2 etc) mais sans land_use_change_co2.

co2_including_luc = co2 + land_use_change_co2

et

co2 + trade_co2 = consumption_co2

```python
#Je recrée moi-même la colonne de cumul de co2 pour être sûr de bien comprendre son calcul.
tmp = df_france_20.loc[:,['co2','cumulative_co2','cement_co2','cumulative_cement_co2']]
tmp['new_cumul_co2'] = tmp['cumulative_co2']
tmp['new_cumul_co2'] = tmp['new_cumul_co2'].shift(1) + tmp['co2']
tmp
```

```python
#Je regarde si total_ghg est la somme des 3 GES observés : co2, methane et oxyde nitreux
tmp = df_france_20.loc[:,['country','year', 'co2_including_luc', 'co2', 'land_use_change_co2', 'methane','nitrous_oxide' , 'consumption_co2','total_ghg', 'total_ghg_excluding_lucf']]
tmp['total_ghg_custom'] = tmp['co2_including_luc'] + tmp['methane'] + tmp['nitrous_oxide']
tmp['total_ghg_excluding_lucf_custom'] = tmp['co2'] + tmp['methane'] + tmp['nitrous_oxide']
tmp['diff_ghg'] = tmp['total_ghg'] - tmp['total_ghg_custom']
tmp['diff_ghg_excluding_lucf'] = tmp['total_ghg_excluding_lucf'] - tmp['total_ghg_excluding_lucf_custom']
display(tmp)
```

On se rend compte qu'il y a une différence entre total_ghg et la somme des 3 GES observés (co2, methane, oxyde nitreux) => les données ne proviennent pas des mêmes sources (d'après le détail des sources sur le site owid-data), ce qui explique ces différences (méthodologies pour les calculs et observations qui diffèrent).
